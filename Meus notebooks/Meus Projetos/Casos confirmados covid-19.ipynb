# CASOS COVID

```python
import pandas as pd
from urllib.request import urlretrieve
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
```

```python
urlretrieve(url, 'global_cases_covid19.csv')
```

```python
df_covid=pd.read_csv('global_cases_covid19.csv')
```

```python
df_covid.head()
```

```python
#excluir latitude e longitude
df_covid.drop(['Lat','Long'],axis = 1, inplace=True)
```

```python
df_country=df_covid.groupby(by='Country/Region').sum()
```

```python
df_country
```

```python
df_country.loc['Brazil']
```

```python
date = df_country.loc['Brazil'].index
cases = df_country.loc['Brazil'].index
```

```python
plt.figure(figsize=(18,10))
plt.bar(date, cases)
```

```python
s_brazil=df_country.loc['Brazil']
```

```python
s_brazil
```

```python
s_brazil=s_brazil[s_brazil > 0]
```

```python
s_brazil.index
```

```python
s_brazil.values
```

```python
plt.figure(figsize=(14,10))
plt.bar(s_brazil.index,s_brazil.values)
```

```python
#somente os ultimos 10 dias

tam = len(s_brazil)

plt.figure(figsize=(14,10))
plt.xticks(rotation=60)
plt.bar(s_brazil.index[tam-30:tam],s_brazil.values[tam-30:tam],linewidth=5)
plt.title("Casos covid ultimos 30 dia no Brasil")
```
