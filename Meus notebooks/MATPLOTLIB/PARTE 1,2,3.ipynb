```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x * x
```

```python
y 
```

```python
# Funcional
# Forma mais tradicional e rapida de plotar informações de duas variaveis (sem muito controle)
plt.plot(x,y, 'r--')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title("Titulo")
```

```python
plt.subplot(1,2,1)#terceiro elemento indica com qual você trabalha
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g-*')
```

```python
fig = plt.figure()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])

axes1.plot(x,y)
axes1.set_xlabel("eixo x")
axes1.set_title("Titulo")
axes2.plot(y,x)
```

```python
fig,ax=plt.subplots()
ax.plot(x,x**3,'b--')
ax.plot(x,x**4,'g.-')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title("Titulo")
```
