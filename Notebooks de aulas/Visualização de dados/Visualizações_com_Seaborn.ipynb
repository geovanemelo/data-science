## Aplicando o estilo do Seaborn no Matplotlib

O Seaborn fornece uma interface mais *high-level* para criarmos gráficos elaborados, visualmente melhores e mais complexos. É muito comum vermos *notebooks* de *Data Science* usando `seaborn` como biblioteca primária para plotar gráficos.

Entretanto, mesmo quem prefere usar o `matplotlib` consegue tirar proveito do estilo que está por trás do `seaborn`.

```python
# importando pacotes

import numpy as np
import matplotlib.pyplot as plt
```

```python
# gerar dados aleatórios

x = np.arange(100)
y = np.random.random_integers(-50,50,100)

# plotando com matplotlib

fig,ax = plt.subplots(figsize=(8,4))
ax.plot(x,y)

plt.tight_layout()
```

O gráfico acima é bem no estilo que estamos acostumados até então aqui no curso. Mas veja só o que é possível melhorar apenas importando a biblioteca `seaborn` e chamando o método `sns.set()`.

```python
#importando seaborn
import seaborn as sns

sns.set()

#plotar gráfico com matplotlib usando estilo do seaborn

fig,ax = plt.subplots(figsize=(8,4))
ax.plot(x,y)

plt.tight_layout()
```

#Scatter plot

Retrata a distribuiçao conjunta de duas variavéis usando uma "nuvem de pontos", onde cada ponto representa um observação no dataset

Há dois tipos distintos de plotar o scatter plat no seaborn. O mais básico deve ser usado **quando ambas variaveis são numéricas**, é o `scatterplot()`. O scatterplot() é padrão em `relplot()`


```python
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill",y="tip",data=tips);
```

Após postados os pontos em duas dimensões, uma outra dimensão pode ser adicionada no plot, colorindo os pontos de acordo com uma terceira variavél. Em seaborn, ela é referida como `hue`

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",data=tips);
```

Para enfatizar a diferenças entre as duas classes, podemos usar um diferente estilo de marcação p/ cada class

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='time',data=tips);
```

nos exemplos acima, o `hue` eram categoricos, então o padrão *qualitative palett* foi aplicado. Se o `hue` é numerico, a coloração padrão será substituida para uma paleta sequencial.

```python
sns.relplot(x="total_bill",y='tip',hue='size',data=tips);
```

nos dois casos, você pode customizar a paleta de cors. Existem varias opçoes. Aqui, irei customizar com a paletta `cubehelix_palett()`

```python
sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-.9,l=.75", data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:start=2.8,rot=.1", data=tips);
```

O terceiro tipo de semantica variavel, muda o tamanho dos pontos, podendo ser usada junta com o hue

```python
sns.relplot(x='total_bill',y='tip',size='size',hue='size',data=tips);
```

ao contrario do `matplotlib.pyplot.scatter()`, o valor literal da variavél não é usado para escolher a areea do ponto. Em vez disso, o *range* dos valores em é usado para areas unitarias. Esse *range* pode ser customizado.

```python
sns.relplot(x='total_bill',y='tip',size='size',sizes=(15,200),data=tips,hue='size',palette="ch:r=-.9,l=.75");
```

#continuando com plots de linhas

Scatter plots são muito eficazes, mas não é o tipo ideal para todas as visualizações. A representação visual deve ser adaptada p/ as especificações do dataset e a questão que você está tentando responder com o plot

Em alguns datasets, você pode querer entender as mudanças em uma variaveis como uma função de tempo, ou outra similar variavel continua. Nessa situação, uma boa escolha é desenhar uma plot de linha. No seaborn, ele pode ser realziado com o `lineplot()`  de forma direta ou com `relplot()`com o "kind='line'"

```python
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x='time',y='value',kind='line',data=df)
g.fig.autofmt_xdate()
```

#Agregando e representando incerteza

O mais complexos datatasets terão multiplas medidas do mesmo valor na variavel 'x'. O comportamento padrão no seaborn é para agregar as medidas multiplas em cada valor x, plotando a média e o intervalo de 95% em torno da media.

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri)
```

Os intervalos de confiança são computados usando bootstrapping, qual pode ser demorado em datasets com muitos dados. É portanto possivel desabilitar

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri,ci=None)
```

Uma outra boa opçao, especialmente com grandes dados, é representar a propagação de distribuiçao em cada ponto de tempo plotando o desvio padrão invez de do intervalo de confiança

```python
sns.relplot(x="timepoint", y="signal", kind="line", ci="sd", data=fmri);
```

Para desligar a agregação completamente, set o parametro `estimator` para `none`. Isso pode causar um estranho efeito quando o dado tem multiplas observation em cada ponto

```python
sns.relplot(x="timepoint", y="signal", estimator=None, kind="line", data=fmri);
```

#plotando subconjutos dos doados com semanticas de mapeamento

A função `lineplot()`  tem a mesma flexibilidade que a `scatterplot()`. Ela pode mostrar até três variaveis modificando o hue,size e stilo dos elementos plotados. E isso é feito usando o mesmo API que o `scatterplot`, signica que não precisamos parar e pensar sobre os paramentros que controlas as linhas e pontos no matplotlib

Usando `lineplot()`, sera tambem determinado como o dato sera agregado. Por exemplo, adicionando a semantica hue com duas divisoes de leveis, o gráfico em duas linhas e bandas de erro, colorindo cada uma para indicar a qual subconjunto dos dados elas correspondem.

```python
sns.relplot(x='timepoint',y='signal',hue='event',kind='line',data=fmri);
```

```python
fmri.head()
```

```python
sns.relplot(x='timepoint',y='signal',hue='region',kind='line',style='event',data=fmri);
```

Você pode identifcar os subsetes com marcadores em cada observação de outras maneiras.


```python
sns.relplot(x='timepoint',y='signal',hue='region',style='event',dashes=False,markers=True,kind='line',data=fmri);
```

como com scatter plots, seja cauteloso ao plotar linhas usando multiplas semanticas. Enquanto algumas vezes informativo, eles podem canusar tambem dificuldades para analisar e interpretar. Mas até quando você esta apenas examinando mudanças em uma variavel adicional, isso pode ser util para alterar cor e estilo de linhhas. Isso pode fazer o plot mais acessivel quando printado em preto e branco ou visto por alguem daltonico

```python
sns.relplot(x='timepoint',y='signal',hue='event',style='event',kind='line',data=fmri);
```

Quando voce está trabalhando com repetidas medidas de datas (isso é, você tem unidades foram mostradas multiplas vezes) você pode tambem plotar cada amostragem de unidade separadamente sem dinstiguir elas através de semanticas. Isso evita desordem na legenda

```python
sns.relplot(x='timepoint',y='signal',hue='region',units='subject',estimator=None,kind='line',data=fmri.query("event == 'stim'"))
```

# Mostrando multiplas relações com facetas

Como tenhhamos enfatizados nesse tutorial, enquanto essas funçoes podem mostrar diversas semanticas, nem sempre são efetivas por si só. Mas e quando você quer que entender a relação entre duass variaveis?

A melhor abordagem pode ser fazer mais de um plot. Devido o `replot()` ser baseado em facegrid, é facil fazer isso. Para mostrar a influencia de uma variavel adicional, inves de atribuir-lo ao um papel semantico no plot, use a face da visualizzação. Isso significa que você pode fazer multiplas figuras e plotar subsets em cada um deles

```python
sns.relplot(x='total_bill',y='tip',hue='time',col='sex',data=tips);
```

```python
tips.head()
```

Você tambem pode mostrar a influência com duas variaveis assim: Uma facetando as linhas e outra facetando as colunas. Como você começa adicionando mais variaveis no grid, você tem pode querer encremetar o tamanho da figura. Lembre-se de que o tamanho do FacetGrid é parametrizado pela altura e proporção de cada faceta:



```python
sns.relplot(x='total_bill',y='tip',hue='sex',data=tips,col='smoker',row='time',height=5);
```

```python
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            col="subject", col_wrap=5,
            height=3, aspect=.75, linewidth=2.5,
            kind="line", data=fmri.query("region == 'frontal'"));
```

No tutorial de enredo relacional vimos como usar diferentes representações visuais para mostrar a relação entre múltiplas variáveis em um conjunto de dados. Nos exemplos, focamos nos casos em que a relação principal estava entre duas variáveis numéricas. Se uma das principais variáveis for "categórica" (dividida em grupos discretos) pode ser útil usar uma abordagem mais especializada para visualização.
